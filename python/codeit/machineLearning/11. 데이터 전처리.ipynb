## 05. scikit-learn으로 과적합(overfitting) 문제 직접 보기

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures

from math import sqrt

import numpy as np
import pandas as pd
```

```python
ADMISSION_FILE_PATH = 'data/admission_data.csv'
admission_df = pd.read_csv(ADMISSION_FILE_PATH).drop('Serial No.', axis=1)
```

```python
admission_df
```

```python
X = admission_df.drop(['Chance of Admit '], axis=1)

polynomial_transformer = PolynomialFeatures(6) # 차원 변형기 정의 (6차항의 다항 회귀 모델 생성)
polynomial_features = polynomial_transformer.fit_transform(X.values)
features = polynomial_transformer.get_feature_names(X.columns)

X = pd.DataFrame(polynomial_features, columns=features)
X
```

```python
y = admission_df[['Chance of Admit ']]
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_train_predict = model.predict(X_train)
y_test_predict = model.predict(X_test)
```

```python
mse = mean_squared_error(y_train, y_train_predict)

print('training set에서의 성능')
print('-----------------------')
print(sqrt(mse))  #0.001504807630434823 (닞은 오차)

mse = mean_squared_error(y_test, y_test_predict)

print('test set에서의 성능')
print('-----------------------')
print(sqrt(mse))  #5.090721963390173 (높은 오차)   ==> 과적합
```

## 10. scikit-learn으로 과적합(overfitting) 문제 해결하기

```python
from sklearn.linear_model import Lasso # Lasso 모델 정의
```

```python
model = Lasso(alpha=0.001, max_iter=1000, normalize=True)
model.fit(X_train, y_train)
```

```python
y_train_predict = model.predict(X_train)
y_test_predict = model.predict(X_test)
```

```python
mse = mean_squared_error(y_train, y_train_predict)

print('training set에서의 성능')
print('-----------------------')
print(sqrt(mse))  #0.001504807630434823 (닞은 오차)

mse = mean_squared_error(y_test, y_test_predict)

print('test set에서의 성능')
print('-----------------------')
print(sqrt(mse))  #0.060077190926892544 (test set에서도 낮은 오차를 보임 ==> 과적합 방지)
```

## 03. scikit-learn으로 k겹 교차 검증 해보기

```python
from sklearn import datasets
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import numpy as np
import pandas as pd

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning) # 경고메시지
```

```python
iris_data = datasets.load_iris() # 붓꽃 데이터 불러오기
```

```python
X = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
y = pd.DataFrame(iris_data.target, columns=['Class'])
```

```python
logistic_model = LogisticRegression(max_iter=2000) # 모델 생성
```

```python
np.average(cross_val_score(logistic_model, X, y.values.ravel(), cv=5)) # k교차 검증(k=5) / 리턴값: 각 데이터셋에 대한 성능 ==> 평균으로 출력
```

## 07. scikit-learn으로 그리드 서치 해보기

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import GridSearchCV  # 그리디서치
from math import sqrt

import numpy as np
import pandas as pd
```

```python
# 데이터 준비
ADMISSION_FILE_PATH = 'data/admission_data.csv'
admission_df = pd.read_csv(ADMISSION_FILE_PATH)

X = admission_df.drop('Chance of Admit ', axis=1)

polynomial_transformer = PolynomialFeatures(2) # 2차식 변형기를 정의
polynomial_features = polynomial_transformer.fit_transform(X.values)

features = polynomial_transformer.get_feature_names(X.columns)

X = pd.DataFrame(polynomial_features, columns=features)
y = admission_df[['Chance of Admit ']]
```

```python
hyper_parameter = {
    'alpha': [0.01, 0.1, 1, 10],
    'max_iter': [100, 500, 1000, 1500, 2000]
}
```

```python
lasso_model = Lasso()
```

```python
hyper_parameter_tuner = GridSearchCV(lasso_model, hyper_parameter, cv=5)
hyper_parameter_tuner.fit(X, y)
```

```python
hyper_parameter_tuner.best_params_
```

```python
def Geni(class_1, class_2):
    class_1_per = class_1 / (class_1 + class_2)
    class_2_per = class_2 / (class_1 + class_2)

    return 1 - (class_1_per ** 2) - (class_2_per ** 2)
print(Geni(10, 40))
print(Geni(60, 15))
```
