```python
# 06. 카테고리 분류
df = pd.read_csv('data/laptops.csv')
df.head()
```

```python
brand_nation = {
    'Dell' : 'U.S.',
    'Apple' : 'U.S.',
    'Acer' : 'Taiwan',
    'HP' : 'U.S.',
    'Lenovo' : 'China',
    'Alienware' : 'U.S.',
    'Microsoft' : 'U.S.',
    'Asus' : 'Taiwan'
}
```

```python
df['brand'].map(brand_nation) #파이썬 객체의 우측 내용으로 데이터 변경
```

```python
df['brand nation'] = df['brand'].map(brand_nation)
df
```

```python
nation_groups = df.groupby('brand nation') #카테고리 비교
type(nation_groups)
```

```python
nation_groups.count()
```

```python
nation_groups.max()
```

```python
nation_groups.first()
```

```python
nation_groups.plot(kind='box', y='price')
```

```python
nation_groups.plot(kind='hist', y='price')
```

```python
# 데이터 합치기
price_df = pd.read_csv('data/vegetable_price.csv')
quantity_df = pd.read_csv('data/vegetable_quantity.csv')
```

```python
price_df
```

```python
quantity_df
```

```python
pd.merge(price_df, quantity_df, on='Product') # inner join
```

```python
pd.merge(price_df, quantity_df, on='Product', how='left') # letf outer join
```

```python
pd.merge(price_df, quantity_df, on='Product', how='right') # right outer join
```

```python
pd.merge(price_df, quantity_df, on='Product', how='outer') # full outer join
```
